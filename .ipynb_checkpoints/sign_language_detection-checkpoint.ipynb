```python
import cv2
import numpy as np
import os
from matplotlib import pyplot as plt
import time
import mediapipe as mp
```

```python
mp_holistic = mp.solutions.holistic # this is the holistic model
mp_drawing = mp.solutions.drawing_utils # this are the draing utilities 


# we are going to use mp+holistocs to make the detections  and mp_drawing to draw them
```

```python
def mediapipe_detection(image, model): # so in order to make the detection we first have to take the image which is in bgr and convert them to rgb and the make it to un writable for us to save some memories and after which we make the detection on the feed and then we convert ot back to writable and then from rgb to bgr 
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image.flags.writeable = False
    result = model.process(image)
    image.flags.writeable = True
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    return image,result
```

```python
def draw_landmarks(image ,results): # we are grabing the results which are the landmaeks ans actually rendering it 
    mp_drawing.draw_landmarks(image,results.face_landmarks, mp_holistic.FACEMESH_TESSELATION)
    mp_drawing.draw_landmarks(image,results.pose_landmarks, mp_holistic.POSE_CONNECTIONS)
    mp_drawing.draw_landmarks(image,results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS)
    mp_drawing.draw_landmarks(image,results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS)
    
    
```

```python
def draw_styled_landmarks(image, results): # It is basically the same thing as the above function but here only we have done some formatting and you could simpily do the changes in the above one 
    # Draw face connections
    mp_drawing.draw_landmarks(image, results.face_landmarks, mp_holistic.FACEMESH_TESSELATION,
                              mp_drawing.DrawingSpec(color=(80,110,10), thickness=1, circle_radius=1),
                              mp_drawing.DrawingSpec(color=(80,256,121), thickness=1, circle_radius=1))
    
    # Draw pose connections
    mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_holistic.POSE_CONNECTIONS,
                              mp_drawing.DrawingSpec(color=(80,22,10), thickness=2, circle_radius=4),
                              mp_drawing.DrawingSpec(color=(80,44,121), thickness=2, circle_radius=2))
    
    # Draw left hand connections
    mp_drawing.draw_landmarks(image, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS,
                              mp_drawing.DrawingSpec(color=(121,22,76), thickness=2, circle_radius=4),
                              mp_drawing.DrawingSpec(color=(121,44,250), thickness=2, circle_radius=2))
    
    # Draw right hand connections
    mp_drawing.draw_landmarks(image, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS,
                              mp_drawing.DrawingSpec(color=(245,117,66), thickness=2, circle_radius=4),
                              mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2))
```

```python
def extract_keypoint(result):
    # this is the simplified version of the logic above 
    pose = np.array([[res.x, res.y, res.z, res.visibility] for res in result.pose_landmarks.landmark]).flatten() if result.pose_landmarks else np.zeros(132)
    lh = np.array([[res.x, res.y, res.z] for res in result.left_hand_landmarks.landmark]).flatten() if result.left_hand_landmarks  else np.zeros(21*3)
    rh = np.array([[res.x, res.y, res.z] for res in result.right_hand_landmarks.landmark]).flatten() if result.right_hand_landmarks  else np.zeros(21*3)
    face = np.array([[res.x, res.y, res.z] for res in result.face_landmarks.landmark]).flatten() if result.face_landmarks  else np.zeros(1404)
    return np.concatenate([pose,face,lh,rh])
        
```

```python
DATA_PATH = os.path.join('MP_Data') # path for the exported data which is all the numpy arrays
actions = np.array(['hello', 'thankyou','iloveyou']) # array of all the actions we want to detect
no_sequence = 30 # 30 videos worth of data
sequence_lenght = 30 # 30 frames in lenght 
```

```python
for action in actions:
    for sequence in range(no_sequence):
        try:
            os.makedirs(os.path.join(DATA_PATH, action , str(sequence)))
        except:
            pass
```

collect keypoint values for training and testing


```python
cap = cv2.VideoCapture(0) # we are initializing a cap variable where we are reading the webcame feed and the videocapture function is used to capture the video from the webcam which is generally 0 
#set mediapipe model
with mp_holistic.Holistic(min_detection_confidence=0.5,min_tracking_confidence=0.5) as holistic:
    #loop through the number of actions 
    for action in actions:
        # looping through number of sequence that is number of videos
        for sequence in range(no_sequence):
            #loop through lenght of videos 
            for frame_num in range(sequence_lenght):
                ret, frame = cap.read() 
                #make detections 
                image ,result = mediapipe_detection(frame,holistic)
                print(result)
                #draw landmarks 
                draw_styled_landmarks(image,result)
                # Apply collection logic
                if frame_num == 0:
                    cv2.putText(image, 'STARTING COLLECTION', (120,200), 
                                cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), 1, cv2.LINE_AA)
                    cv2.putText(image, 'Collecting frames for {} Video Number {}'.format(action, sequence), (15,12), 
                                cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,255), 1, cv2.LINE_AA)
                    cv2.waitKey(2000)
                else:
                    cv2.putText(image, 'Collecting frames for {} Video Number {}'.format(action, sequence), (15,12), 
                                cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,255), 1, cv2.LINE_AA)
                # new keypoint extractiom
                keypoints = extract_keypoint(result)
                npy_path = os.path.join(DATA_PATH, action, str(sequence), str(frame_num))
                np.save(npy_path,keypoints)
                # Show to screen
                cv2.imshow('OpenCV Feed', image) 
                #break capturing 
                if cv2.waitKey(10) & 0xFF == ord('q'): 
                    break
    cap.release() # here we just relese the webcam which is just stop the webcam 
    cv2.destroyAllWindows()
```

preprocessing the data and creating the labels 


```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
```

```python
label_map = {label:num for num , label in enumerate(actions)}
```

```python
sequences,labels = [] ,[]
for action in actions:
    for sequence in range(no_sequence):
        window = []
        for frame_num in range(sequence_lenght):
            res = np.load(os.path.join(DATA_PATH, action , str(sequence), "{}.npy".format(frame_num)))
            window.append(res)
        sequences.append(window)
        labels.append(label_map[action])
```

```python
x = np.array(sequences)
```

```python
y = to_categorical(labels).astype(int)
```

```python
y
```

```python
X_train , X_test, y_train , y_test = train_test_split(x,y, test_size=0.05)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM , Dense
from tensorflow.keras.callbacks import TensorBoard
```

```python
log_dir = os.path.join('logs')
tb_callback = TensorBoard(log_dir=log_dir)
```

```python
model = Sequential()
model.add(LSTM(64, return_sequences=True, activation='relu', input_shape=(30, 1662)))
model.add(LSTM(128, return_sequences=True, activation='relu'))
model.add(LSTM(64, return_sequences=False, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(actions.shape[0], activation='softmax'))
```

```python
model.compile(optimizer='Adam' , loss = 'categorical_crossentropy' , metrics=['categorical_accuracy'])
```

```python
model.fit(X_train,y_train,epochs=1000, callbacks =[tb_callback])
```
